Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = [1,2,3,4,5,6,7]
y = [28,96,54,29,56,9,25]
```

```python
plt.plot(x,y, color = 'green', linewidth = 5, linestyle = 'dashdot')
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('Weather data')
plt.plot(x,y, color='red', linewidth = 10, linestyle='dotted')
```

Axes Labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather data')

plt.plot(days, max_t, label = 'max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

plt.legend()
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather data')

plt.plot(days, max_t, label = 'max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

plt.legend(loc = 'lower left', shadow=True)
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather data')

plt.plot(days, max_t, label = 'max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

plt.legend(loc = 'best', shadow=True, fontsize = 'large')
plt.grid()
```

Bar Chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('company data')
plt.grid()
plt.bar(company, revenue)
```

```python
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('company data')
plt.grid()
plt.barh(company, revenue)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.ylabel('revenue')
plt.title('company data')
#plt.grid()
plt.xticks(xpos, company)
plt.bar(xpos, revenue, label='revenue')
plt.legend(loc = 'best')
```

```python
profit=[40,2,34,12]
```

```python
plt.ylabel('revenue')
plt.title('company data')
#plt.grid()
plt.xticks(xpos, company)
plt.bar(xpos, revenue, label='revenue')
plt.bar(xpos, profit, label = 'profit')
plt.legend(loc = 'best')
```

```python
plt.ylabel('revenue')
plt.title('company data')
#plt.grid()
plt.xticks(xpos, company)
plt.bar(xpos-0.2, revenue,width = 0.4, label='revenue')
plt.bar(xpos+0.2, profit,width=0.5, label = 'profit')
plt.legend(loc = 'best')
```

```python
plt.ylabel('revenue')
plt.title('company data')
#plt.grid()
plt.xticks(xpos, company)
plt.barh(xpos, revenue, label='revenue')
plt.barh(xpos, profit, label = 'profit')
plt.legend(loc = 'best')
```

Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar, bins=5, rwidth=0.5)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.5, color='green')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins = [80,100,125,150],color=['g','r'], rwidth=0.8, label = ['men', 'women'])
plt.legend(loc='best', shadow=True)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins = [80,100,125,150],color=['g','r'], rwidth=0.8,histtype='step', label = ['men', 'women'])
plt.legend(loc='best', shadow=True)
```
